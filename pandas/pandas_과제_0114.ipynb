### Q. 10을 곱하는 사용자 함수와 두 객체의 나눗셈을 수행하는 사용자 함수를 정의한 후 결과값이 각각 200과 10을 출력하세요.

```python
def mul_10(x):
    return x*10

def div_ab(a, b):
    try:
        return round(a/b, 2)
    except ZeroDivisionError as e:
        print(e)
        return None
    
print(f'함수 mul_10(x)의 인자값 20 => {mul_10(20)}')
print(f'함수 div_ab(a, b)의 인자값 60, 6 => {div_ab(60, 6)}')
```

### Q. titanic 데이터셋을 seaborn에서 로딩한 후 블린 인덱싱을 이용하여 아래 사항들을 수행하세요.
- 나이가 10대(10~19세)인 승객만 따로 선택
- 나이가 10세 미만(0~9세)이고 여성인 승객만 따로 선택
- 나이가 10세 미만(0~9세) 또는 60세 이상인 승객의 age, sex, alone 열만 선택

```python
# 나이가 10대(10~19세)인 승객만 따로 선택

teen = titanic.loc[(titanic['age']>=10) & (titanic['age']<20)]
display(teen)
```

```python
# 나이가 10세 미만(0~9세)이고 여성인 승객만 따로 선택

kid_f = titanic.loc[(titanic['age']<10) & (titanic['sex']=='female')]
display(kid_f)
```

```python
# 나이가 10세 미만(0~9세) 또는 60세 이상인 승객의 age, sex, alone 열만 선택

kid_elder = titanic.loc[(titanic['age']<10) | (titanic['age']>=60), ['age', 'sex', 'alone']]
display(kid_elder)
```

### Q. 각 열의 NaN 찾기, 각 열의 NaN 개수 반환, 데이터프레임의 총 NaN 개수를 반환하는 사용자 함수를 정의한 후 df.pipe 메소드로 함수 매핑한 결과를 모두 출력하세요.

```python
def findnan(a):
    return a.isnull()
def sumnan(a):
    return a.isnull().sum()
def sumnanall(a):
    return a.isnull().sum().sum()

df2 = df.copy()
display(df2.pipe(findnan))
print()
print(df2.pipe(sumnan), '\n')
print(df2.pipe(sumnanall))
```

### Q. 타이타닉 df의 열 순서를 아래와 같이 재구성하고 출력하세요
* 열 선택 : ['pclass', 'sex', 'age', 'survived'].
* 열 이름의 리스트 만들기
* 열 이름을 알파벳 순으로 정렬하기
* 열 이름을 기존 순서의 정반대 역순으로 정렬하기
* 열 이름을 사용자가 정의한 임의의 순서로 재배치하기

```python
# 열 선택 : ['pclass', 'sex', 'age', 'survived'].

tdf = pd.DataFrame(titanic, columns=['pclass', 'sex', 'age', 'survived'])
display(tdf)
```

```python
# 열 이름의 리스트 만들기

col = list(tdf.columns)
print(col)
```

```python
# 열 이름을 알파벳 순으로 정렬하기

col_sorted = sorted(col)
print(col_sorted)
```

```python
# 열 이름을 기존 순서의 정반대 역순으로 정렬하기

# print(col[::-1])
col_reversed = list(reversed(col))
print(col_reversed)
```

```python
# 열 이름을 사용자가 정의한 임의의 순서로 재배치하기

col_customed = ['survived', 'pclass', 'age', 'sex']
tdf_customed = tdf[col_customed]
display(tdf_customed)
```
