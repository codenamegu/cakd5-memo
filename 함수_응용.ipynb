```python
def factorial(n):
    output = 1
    for i in range(1, n+1):
        output *= i
    return output

x = int(input("숫자 입력: "))
print(f'{x}! = {factorial(x)}')
```

```python
# 재귀란 '자기 자신을 호출하는 것'을 의미
# 재귀함수를 이용해서 팩토리얼을 구하는 사용자 함수를 작성하고 1!, 5!, 10! 구하기
# factorial(n) = n* factorial(n-1) (n>=1)

def factorial(n):
    if(n==0):
        return 1
    else:
        return n*factorial(n-1)
    
x = int(input("숫자 입력: "))
print(f'{x}! = {factorial(x)}')
```

```python
# 연습문제 - 원의 면적과 둘레의 길이를 구하는 사용자 함수를 작성하여 반지름이 5와 10일 때 면적과 둘레

def getEx(r):
    return r*r*3.14
def getLen(r):
    return 2*r*3.14
    
x = int(input("원의 반지름: "))
print(f'면적: {getEx(x):.2f}')
print(f'둘레: {getLen(x):.2f}')
```

```python
# 연습문제 - 문자열 리스트 states를 정형화하여 아래와 같이 출력
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina', 'West virginia']
# [출력]
# ['Alabama',
#  'Georgia',
#  'Georgia',
#  'Georgia',
#  'Florida',
#  'South   Carolina',
#  'West Virginia']

def clean_strings(strings):
    result = []
    for value in strings:
        value = value.strip()
        result.append(value)
    return result
clean_strings(states)
```

```python
# [도전과제] - 문자열 리스트 states를 정형화하여 아래와 같이 출력
import re # 정규표현식 패키지

states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina##', 'West virginia?']
# [출력]
# ['Alabama',
#  'Georgia',
#  'Georgia',
#  'Georgia',
#  'Florida',
#  'South   Carolina',
#  'West Virginia']

for i, value in enumerate(states):
    value = value.title().strip() # 첫글자 대문자화 / 빈칸 제거
    value = re.sub('[!#?]', '', value) # 특수문자 제거(''로 대체)
    states[i] = value
    print(value)
    
```

```python
# [과제] 1000명에 대하여 아래와 같이 출력
# BMI = round(int(weight) / (int(height) / 100) ** 2), 2)
# 이름 : 라바
# 몸무게 : 80
# 키 : 146
# BMI : 35.56789
# 결과: 과체중

import random
hanguls = list('가나다라마바사아자차카타파하')
with open('info.txt','w') as file:
    for i in range(1000):
        name = random.choice(hanguls) + random.choice(hanguls)
        height = random.randrange(140,200)
        weight = random.randrange(40,100)
        bmi = round(int(weight) / (int(height)/100)**2)
        cons = ''
        if bmi < 20:
            cons = '저체중'
        elif bmi >= 20 and bmi <25:
            cons = '정상'
        elif bmi >= 25 and bmi <30:
            cons = '과체중'
        else:
            cons = '비만'

        file.write(f'{name}, {height}, {weight}, {bmi}, {cons}\n')
with open('info.txt', 'r') as file:
    for line in file:
        (name,weight,height,bmi,cons) = line.strip().split(',')
        print('\n'.join([
            "이름  : {}",
            "몸무게: {}",
            "키    : {}",
            "BMI   : {}",
            "결과  : {}"
            ]).format(name,weight,height,bmi,cons))
        print()
```

```python
#

def getTextFreq(filename):
    with open(filename, 'r') as f:
        text = f.read()
        fa = {}
        for c in text:
            if c in fa:
                fa[c]+=1
            else:
                fa[c]=1
    return fa

ret = getTextFreq('data/mydata.txt')
ret = sorted(ret.items(), key=lambda x:x[1], reverse=1)
print(ret)
```

```python
# [도전과제] 연도 / 성별 / 출생아 수
# 1880, 90994, 110490

year=1880
while year<2016:
    filename = 'data/names/yob' + str(year) + '.txt'
    with open(filename) as file:
        f_sum=0
        m_sum=0
        for line in file:
            (name, gender, birth) = line.strip().split(',')
            if(gender == 'F'):
                f_sum+=int(birth)
            elif(gender == 'M'):
                m_sum+=int(birth)
        print(f'{year}년도\n\t여자아이 > {f_sum}명\n\t남자아이 > {m_sum}명')
    year+=1
```

```python
from os.path import exists

# 사용자 함수 정의
def getTop10BabyName(year):
    nameF = {}
    nameM = {}
    
    filename = 'data/names/yob%s.txt'%year
    if not exists(filename):
        print('[%s] 파일이 존재하지 않습니다.'%filename)
        return None # 프로그램 바로 종료
    
    with open(filename,'r') as f:
        data = f.readlines()
        for d in data:
            if d[-1] == '\n':
                d = d[:-1]
                
                tmp = d.split(',')
                name = tmp[0]
                sex = tmp[1]
                birth = tmp[2]
                
                if sex == 'F':
                    ret = nameF
                else:
                    ret = nameM
                    
                if name in ret:
                    ret[name] += int(birth)
                else:
                    ret[name] = int(birth)
    
    # 정렬
    retF = sorted(nameF.items(), key=lambda x:x[1],reverse=True)
    retM = sorted(nameM.items(), key=lambda x:x[1], reverse=True)

    for i, name in enumerate(retF):
        if i>9:
            break
        print('TOP_%d 여자아기이름:%s'%(i+1,name))
    for i, name in enumerate(retM):
        if i>9:
            break
        print('Top_%d 남자아기이름:%s'%(i+1,name))
        
y = input('출생년도 입력하세요(예:2001):')
getTop10BabyName(y)
```

```python
import time

def long_task(): # 5초의 시간이 걸리는 함수
    for i in range(5):
        time.sleep(1) # 1초간 대기
        print(f'working : {i}')

print('Start')
for i in range(5):
    long_task()
print('End')
```

```python
import time
import threading

def long_task():
    for i in range(5):
        time.sleep(1)
        print(f'working : {i}')
        
print('Start')

threads = []
for i in range(5):
    t = threading.Thread(target=long_task)
    threads.append(t)
    
for t in threads:
    t.start()
    
for t in threads:
    t.join() # join으로 스레드가 종료될 때까지 기다린다.
    
print('End')
```
