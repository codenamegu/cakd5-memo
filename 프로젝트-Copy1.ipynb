```python
# 정보 추가

import cryptocode as cpt
import pickle

def newUser():
    file = open('UserPW.txt', 'ab')
    
    print("사용자 정보를 등록합니다.")
    
    while 1:
        try:
            ID=input("ID: ")
            PW=input("PW: ")
            name=input("성함: ")
            tel=input("전화번호: ")
            level=int(input("관리 등급: "))
            hint=input("비밀번호 힌트: ")

            data = [['ID', ID], ['PW', PW], ['성함', name], ['전화번호', tel], ['관리 등급', level], ['비밀번호 힌트', hint]]
            print('\n'+'='*30)
            for line in data:
                print(f'{line[0]} : {line[1]}')
            print('='*30)

            while 1:
                print()
                try:
                    answer=input("입력하신 정보로 사용자 등록을 하시겠습니까?[y/n] > ")
                    if(answer=='y'):
                        print("사용자 등록이 완료되었습니다.")
                        subdata={'pw':cpt.encrypt(PW,ID), 'seculevel':level, 'pwhint':cpt.encrypt(hint, ID), 'name':name, 'tel':tel}
                        data = {ID:subdata}
                        pickle.dump(data,file)
                        file.close()
                        return None

                    elif(answer=='n'):
                        select = input("사용자 등록을 취소하시겠습니까?[y/n] > ")
                        if(select == 'y'):
                            print("등록 프로그램을 종료합니다.")
                            file.close()
                            return None
                        elif(select == 'n'):
                            print("정보 입력으로 돌아갑니다.")
                            break
                        else:
                            raise Exception
                except:
                    print("[y] 또는 [n]만 입력해주세요.")
        except KeyboardInterrupt:
            print("등록을 중지합니다. 메인으로 돌아갑니다.")
            file.close()
            return None
    
newUser()
```
